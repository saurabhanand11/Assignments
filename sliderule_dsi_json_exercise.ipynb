## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
import pandas as pd
```

```python
import json
from pandas.io.json import json_normalize
```

```python
from numpy import nan
```

# Question 1- Find the 10 countries with most projects

```python
df1=pd.read_json('data/world_bank_projects.json')['countryname'].value_counts().reset_index()[:10]
df1.columns=['countryname','Project_count']
df1.head(10)
```

# Question 2- Find the top 10 major project themes 

```python
# load json as string
df_theme=json.load((open('data/world_bank_projects.json')))
themes=json_normalize(df_theme, 'mjtheme_namecode')

```

```python
#Get the count of the code
codecnt=themes['code'].value_counts()
codecnt.name='counts'
codecnt.columns=['code','Theme_count']
codecnt
```

```python
#Display code name
codename=themes.replace('',nan).dropna().groupby('code').last()                                              
codename
```

```python
#Combining above two answers to get a final dataframe
pd.DataFrame([codename['name'],codecnt]).transpose().sort_values('counts',ascending=False)
```

# 3.Question 3- Create a dataframe with the missing names filled in.

```python
themes['name']=themes['code']
themes['name'].replace(to_replace=codename.index.values,value=codename.values,inplace=True)
themes
```
